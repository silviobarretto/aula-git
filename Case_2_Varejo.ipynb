```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
# Agrupando os canais de venda
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
# Realizando o replace de APP e Aplicativo
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
# Agrupando as bandeiras
vendas.groupby('bandeira').idcompra.nunique()
```

```python
# Agrupando os departamentos
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
# Agrupando os estados
vendas.groupby('estado').idcompra.nunique()
```

```python
# Retirando os espao em branco do nomes
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
# Listando os estados com branco
vendas[vendas['estado'].isnull()]
```

```python
# Apagando os estados e preos em branco
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
# A solicitao do cliente  pra substituir os estados em branco por MS
vendas['estado'].fillna('MS', inplace=True) #inplace =  para manter a alterao na base original
vendas.info()
```

```python
# Criando uma mdia de preos para preencher os campos de preos em branco
media_preco = vendas['Preo'].mean()

vendas['Preo'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas.head()
```

```python
vendas[['Preo','Preo_com_frete']].describe()
```

```python
preco_errado = vendas.query('Preo > Preo_com_frete')

vendas_correto = vendas.query('Preo < Preo_com_frete')
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
# Departamentos mais vendidos
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# Mdia de preo com frete por Nome de Departamento
# Round = para definir qtas casas decimais round(,2)
round(vendas_correto.groupby('Nome_Departamento')['Preo_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
# Quantidade de venda por ms, aqui est em ano
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# Criando uma coluna ms para ver a qtd de vendas ms
vendas_correto['mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()

vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
clientes = pd.read_excel('cliente_varejo.xlsx')
clientes.head()
```

```python
vendas_correto.head()
```

```python
clientes[['idade', 'renda']].describe()
```

```python
clientes.info()
```

```python
clientes = clientes.astype({'renda':'float'})
clientes.info()
```

```python
# Unindo as tabelas (Join)
vendas_clientes = vendas_correto.merge(clientes, how='left', on='cliente_Log')
vendas_clientes.head()
```

```python
# Mdia de renda para cada tipo de canal de venda
agg_idcanal_renda = round(vendas_clientes.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
# Mdia de idade de clientes por bandeira
agg_bandeira_idade = round(vendas_clientes.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_bandeira_idade
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_bandeira_idade['bandeira'], agg_bandeira_idade['idade'], color='green')
plt.ylabel('Mdia de idade')
plt.xlabel('Bandeira')
plt.title('Mdia de idade de clientes por bandeira', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='green')
plt.ylabel('Mdia de renda')
plt.xlabel('Canal de venda')
plt.title('Mdia de renda por canal de venda', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
venda_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(15,6))
plt.xlabel('Data da venda')
plt.ylabel('Quantidade de vendas')
x = venda_por_data['Data']
y = venda_por_data['idcompra']

plt.plot(x, y, color='green', linewidth='2')

plt.show()
```

```python
agg_depto_preco = round(vendas_correto.groupby('Nome_Departamento')['Preo_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_depto_preco
```

```python
fig = plt.figure(figsize=(40,20))
plt.bar(agg_depto_preco['Nome_Departamento'], agg_depto_preco['Preo_com_frete'], color='green')
plt.ylabel('Mdia de Preo com frete')
plt.xlabel('Nome do departamento')
plt.title('Mdia de Preo com frete por Departamento', loc='center', fontsize=18, fontweight='bold', color='black');
```
